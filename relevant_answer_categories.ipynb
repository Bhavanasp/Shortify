```python
import pandas as pd
```

```python
rel_que = pd.read_csv('data/relevant_questions.csv')
ans = pd.read_csv('data/keyword_answer.csv')
kdf = pd.read_csv('data/keywords.csv')
```

```python
rel_que
```

```python
ans['question'] = ans
```

```python
kdf
```

```python
rel_ans = pd.DataFrame()
```

```python
for i in enumerate(rel_que['id']):
    rel_ans = pd.concat([ans[ans['id'] == i[1]], rel_ans])
```

```python
rel_ans
```

```python
topic_ans = pd.DataFrame()
```

```python
topic_qa = [[],[],[],[],[],[],[],[]] 
```

```python
for i in range(0, rel_ans.shape[0]):
    topic_list = rel_ans.iloc[i]['topic list'].split(',')
    for topic in topic_list:
        topic = int(topic.split(' ')[-1])
        que_row = rel_que[rel_que['id']==rel_ans['id'].iloc[i]]
        que = que_row.iloc[0]['questions']
        ans = rel_ans.iloc[i]['answer']
        topic_qa[topic].append([que, ans])
```
